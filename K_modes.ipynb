```
!pip install kmodes
import pandas as pd
from sklearn import preprocessing
from kmodes.kmodes import KModes
label_encoder = preprocessing.LabelEncoder()
```

```
Konsumen = {'Cabang': ['Pekanbaru','Pekanbaru','Medan','Jakarta','Jakarta','Jakarta','Medan','Pekanbaru','Pekanbaru','Medan','Medan','Medan'],
            'Domisili': ['Pekanbaru','Pekanbaru','Medan','Jakarta','Jakarta','Tanggeran','Medan','Pekanbaru','Pekanbaru','Medan','Medan','Binjai'],
            'Nama': ['Gila','Gilang','Gilang','Haykal','Haykal','Haykal','Haykal','Agng','Agung','Agung','Dimas','Dimas'],
            'Cabang': ['Pekanbaru','Pekanbaru','Medan','Jakarta','Jakarta','Jakarta','Medan','Pekanbaru','Pekanbaru','Medan','Medan','Medan'],
            'Tempat_Lahir': ['Pekanbaru','Pekanbaru','Pekanbaru','Jakarta','Jakarta','Jakarta','Jakarta','Pekanbaru','Pekanbaru','Pekanbaru','Medan','Medan'],
            'Tahun_Lahir': ['1991','1991','1991','2019','2019','2019','2019','1995','1995','1995','1992','1992'],
            'ID': ['A1991','A1991','B1991','A2019','A2019','A2019','B2019','A1995','A1995','A1995','A1992','A1992'],
            'HP': ['01','01','02','03','03','03','03','04','04','04','05','06'],
            
            }

df = pd.DataFrame(Konsumen, columns = ['Cabang', 'Nama', 'Tempat_Lahir', 'Tahun_Lahir', 'ID','Domisili','HP'])

print (df)
```

```
df_t2 = df
clus_df2 = KModes(n_clusters=4, init = "Huang", n_init = 1000, verbose=1)
fit_clus_df = clus_df2.fit_predict(df_t2)
df_t2["P_Key"] = fit_clus_df
```

```
print(df_t2)
```

```
df_t2.sort_values(by=["P_Key"], inplace=True, ascending=True)
print(df_t2)
```
