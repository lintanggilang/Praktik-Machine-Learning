```
df_Key =  pd.DataFrame()
df_Key["P1"] = df_ml["Nama"]+"#"+df_ml["Tempat_Lahir"]
df_Key["P2"] = df_ml["Nama"]+"#"+df_ml["Tahun_Lahir"]
df_Key["P3"] = df_ml["Nama"]+"#"+df_ml["ID"]
df_Key["P4"] = df_ml["Nama"]+"#"+df_ml["Domisili"]
df_Key["P5"] = df_ml["Nama"]+"#"+df_ml["HP"]
df_Key["P6"] = df_ml["Tempat_Lahir"]+"#"+df_ml["Tahun_Lahir"]
df_Key["P7"] = df_ml["Tempat_Lahir"]+"#"+df_ml["ID"]
df_Key["P8"] = df_ml["Tempat_Lahir"]+"#"+df_ml["Domisili"]
df_Key["P9"] = df_ml["Tempat_Lahir"]+"#"+df_ml["HP"]
df_Key["P10"] = df_ml["Tahun_Lahir"]+"#"+df_ml["ID"]
df_Key["P11"] = df_ml["Tahun_Lahir"]+"#"+df_ml["Domisili"]
df_Key["P12"] = df_ml["Tahun_Lahir"]+"#"+df_ml["HP"]
df_Key["P13"] = df_ml["ID"]+"#"+df_ml["Domisili"]
df_Key["P14"] = df_ml["ID"]+"#"+df_ml["HP"]
df_Key["P15"] = df_ml["Domisili"]+"#"+df_ml["HP"]
```

```
df_Key
```

```
score = []
for k in range(2,10):
    model = KModes(n_clusters=k, init = "Cao", n_init = 1000, verbose=1);
    fit_model = model.fit_predict(df_Key);
    score.append(model.cost_)
```
